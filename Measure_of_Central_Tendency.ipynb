```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
input_data= pd.read_csv('StudentsPerformance.csv')
input_data.head()
```

```python
#Find out how many males and females participated in the test.
int_male =input_data['gender'].tolist().count('female')
int_female=input_data['gender'].tolist().count('male')
print('Male count is:',int_male)
print('Female count is:',int_female)
#Graphical repersentation
plt.title('Gender Comparison ',fontsize=20,fontweight='bold')
gender = ['Male','Female']
count = [int_male,int_female]
plt.bar(gender,count)
plt.show()
```

```python
# What do you think about the students' parental level of education?
input_data['parental level of education'].value_counts()
#Display
plt.figure(figsize=(12,10))
input_data['parental level of education'].value_counts().plot(kind='bar', color= 'green')
plt.title('Student parental level of education',fontsize=20,fontweight='bold')
plt.xlabel('Level',fontsize=15)
plt.ylabel('Count',fontsize=15)
```

```python
# Who scores the most on average for math, reading and writing based on  Gender  Test preparation course
maths_mark=input_data.groupby(['race/ethnicity','gender','test preparation course'],as_index=True).agg({'math score':'max'})
reading_mark=input_data.groupby(['race/ethnicity','gender','test preparation course'],as_index=True).agg({'reading score':'max'})
writing_mark=input_data.groupby(['race/ethnicity','gender','test preparation course'],as_index=True).agg({'writing score':'max'})
print(maths_mark)
print(reading_mark)
print(writing_mark)
```

```python
# What do you think about the scoring variation for math, reading and writing based on  Gender  Test preparation course
print('Analysis of Student Performance')
Min_mark = input_data.groupby(['gender','test preparation course']).min()
print(Min_mark)
Avg_mark = input_data.groupby(['gender','test preparation course']).median()
print(Avg_mark)
Max_mark = input_data.groupby(['gender','test preparation course']).max()
print(Max_mark)
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(y='math score',x='gender',hue='test preparation course',palette = 'Spectral',data=input_data);
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0);
plt.title('Variation in Math Score',fontsize=16,color = 'green');
```

```python
sns.set_context('talk')
sns.set_style('darkgrid')
sns.barplot(y='reading score',x='gender',hue='test preparation course',palette = 'hls',data=input_data);
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0);
plt.title('Variation in Reading Scores',fontsize=20);
```

```python
sns.set_context('talk')
sns.set_style('whitegrid')
sns.barplot(y='writing score',x='gender',hue='test preparation course',data=input_data);
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0);
plt.title('Variation in Writing ',fontsize=18, fontweight = 'bold');
```

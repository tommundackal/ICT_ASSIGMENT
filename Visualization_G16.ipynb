```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Read the Excel data sheet
input_file=pd.read_excel('Telco-Customer-Churn.xlsx')
#Display the Excel sheet
input_file
# Coloumn names
input_file.columns
 
```

```python
# Compare churn count with respect to gender.
input_churn=input_file.groupby('gender').count()
input_churn['Churn']
```

```python
# Find out how many female senior citizens there in the dataset 
data_senior=input_file[input_file['SeniorCitizen']==1]
data_senior_female=data_senior[data_senior['gender']=='Female']
data_senior_female['SeniorCitizen'].count()
```

```python
#Compare 'tenure' with 'Total Charges' 
input_file['tenure'].dtype
input_file['TotalCharges']=pd.to_numeric(input_file['TotalCharges'],errors = 'coerce')
input_file['TotalCharges'].dtype
plt.scatter(input_file['tenure'],input_file['TotalCharges'],s=10,c='green',marker='^')
plt.title('Compare tenure with Total Charges')
plt.xlabel('tenure')
plt.ylabel('Total Charge')
```

```python
#Find out which contract preffered by the senior citizen
data_senior_contract=input_file[input_file['SeniorCitizen']==1].value_counts()
data_senior_contract=data_senior_contract.groupby('Contract').size()
data_senior_contract.plot(kind='bar')
```

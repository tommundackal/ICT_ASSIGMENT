```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
input_data_csv=pd.read_csv('company_sales_data.csv')
#Display the data
#input_data_csv
data_graph=input_data_csv.loc[input_data_csv['facewash']<1500]
#Input required data_graph
data_graph.plot(kind='bar',x='month',y=['facecream','facewash'])
plt.title('Facewash Sale Vs Facecream sales \nLess than for month facewash less 1500')
plt.xlabel('Month')
plt.ylabel('Sales in unit')
```

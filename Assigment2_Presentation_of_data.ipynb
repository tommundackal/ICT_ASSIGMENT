```python
#1.Read the dataset to the python environment
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
input_data_csv=pd.read_csv('cars_data.csv')
#print data of the csv
input_data_csv
#input_data_csv.head()
```

```python
#2.Check for the null values present in the dataset.
input_data_csv.isnull()
#input_data_csv.isna()
```

```python
#3 Plot a bar graph of male vs female buyers participated in the sales.
input_data_graph=input_data_csv[['Buyer Gender','New Car']]
input_data_graph=input_data_graph.groupby(['Buyer Gender','New Car']).size()
input_data_graph=input_data_graph.unstack()
input_data_graph.plot(kind='bar')
plt.title('Male vs Female buyers participated in the sales')
plt.xlabel('Gender')
plt.ylabel('Sales Unit')
```

```python
#4 Find the top 5 cars based on their sales price
input_carmax=input_data_csv.groupby('Model').sum().sort_values('Sale Price', ascending=False).head()
input_carmax
#input_carmax=input_carmax[['Model','Sale Price']]
```

```python
#5 Find the least 5 cars based on their Resell price.
input_least=input_data_csv.groupby('Model').sum().sort_values('Resell Price',ascending=True).tail()
input_least
```
